## Plot the Data - NL

```python
# Line chart showing water levels in a certain province 
# sns.lineplot(data=df_NL)

# Set the width and height of the figure
plt.figure(figsize=(14,6))

# for every day
arr = np.array([20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300, 320, 340, 360, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720])

for a in range(36):
  var = arr[a];
  plt.axvline(x = var, color ='lightblue', linewidth = 0.5)

plt.axhline(y=86.85, color='red')

# Add title
plt.title("Water Levels for Newfoundland for the past 30 days from today")

# x label
plt.xlabel("Data for the past 30 days from today (blue lines indicating a different day and x values indicating hourly values)")

# y label
plt.ylabel("Water level (in m)")

# limits
plt.ylim(top = 86.05)
plt.ylim(bottom = 85.75)

# Line chart 
sns.lineplot(data=df_NL)
```
