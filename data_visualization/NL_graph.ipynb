## Plot the Data - NL

```python
# Line chart showing water levels in a certain province 
# sns.lineplot(data=df_NL)

# Set the width and height of the figure
plt.figure(figsize=(14,6))

# for every day
arr = np.array([24, 48, 72, 96, 120, 144, 168, 192, 216, 240, 264, 288, 312, 336, 360, 384, 408, 432, 456, 480, 504, 528, 552, 576, 600, 624, 648, 672, 696, 720])

for a in range(30):
  var = arr[a];
  plt.axvline(x = var, color ='lightblue', linewidth = 0.5)

plt.axhline(y=86.012, color="#4682B4", linewidth = 0.6)

# Add title
plt.title("Water Levels for Newfoundland for the past 30 days from today", color = 'blue')

# x label
plt.xlabel("Data for the past 30 days from today (blue lines indicating a different day and x values indicating hourly values)")

# y label
plt.ylabel("Water level (in m)")

# limits
plt.ylim(top = 86.05)
plt.ylim(bottom = 85.75)

# Line chart 
sns.lineplot(data=df_NL)
```
